```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np

#Exercício 1

# a. Carregar o dataset
df = pd.read_csv('/kaggle/input/covid-19-clean-complete-csvnias-ia/covid_19_clean_complete.csv')

# b. Quantidade de linhas e colunas
print("Dimensão da base de dados:", df.shape)

# c. Quais são as colunas
print("\nNomes das colunas:")
print(df.columns.tolist())

# d. Tipo de dado de cada coluna
print("\nTipos de dados por coluna:")
print(df.dtypes)

# e. Transformar colunas de data de 'object' para 'datetime64'
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            convertido = pd.to_datetime(df[col], errors='coerce')
            # Se mais de 50% das linhas forem convertidas com sucesso, mantemos como data
            if convertido.notna().mean() > 0.5:
                df[col] = convertido
                print(f"Coluna '{col}' convertida para datetime64")
        except Exception:
            pass

# f. Resumo estatístico do banco de dados
print("\nResumo estatístico:")
print(df.describe(include='all'))

# g. Quais colunas apresentam NaN
print("\nValores nulos por coluna:")
print(df.isnull().sum())
```

```python
# Exercício 2:

# a) Filtrar apenas registros da China
china_df = df[df['Country/Region'] == 'China']

# Agrupar por província, somar casos confirmados e ordenar
provincia_casos = (
    china_df.groupby('Province/State')['Confirmed']
    .sum()
    .sort_values(ascending=False)
)

# Selecionar as 5 províncias com mais casos
top5_provincias = provincia_casos.head(5)
print("Top 5 províncias com mais casos confirmados na China:")
print(top5_provincias, "\n")

# b) Filtrar dados apenas dessas províncias
top5_df = china_df[china_df['Province/State'].isin(top5_provincias.index)]

# c) Manter apenas colunas de interesse
dados_selecionados = top5_df[['Province/State', 'Confirmed', 'Active', 'Deaths', 'Recovered']]

# d) Agrupar novamente por província e somar os valores
resumo_por_provincia = dados_selecionados.groupby('Province/State').sum()
print("Resumo agrupado por província:")
print(resumo_por_provincia, "\n")
print('Função de agrupamento usada: groupby("Province/State").sum() → agrupa e soma valores numéricos por província.\n')

# e) Criar um dataframe final com as 5 regiões e seus dados
top5_regioes_df = top5_df.copy()
print("DataFrame final com as 5 províncias selecionadas:")
print(top5_regioes_df)
```

```python
#Exercício 4:
# a) Importar bases
pop_df = pd.read_csv("/kaggle/input/world-meter-data-csvnias-ia/world meter data.csv")[['Country/Region', 'Continent', 'Population']]
covid_df = pd.read_csv("/kaggle/input/covid-19-clean-complete-csvnias-ia/covid_19_clean_complete.csv")

# Função para combinar país e província
def combinar_pais_provincia(linha):
    if pd.notna(linha['Province/State']):
        return f"{linha['Country/Region']}_{linha['Province/State']}"
    return linha['Country/Region']

# Ajustar nomes de país/província e remover coluna
covid_df['Country/Region'] = covid_df.apply(combinar_pais_provincia, axis=1)
covid_df = covid_df.drop(columns=['Province/State'])

# c) Agrupar por país somando mortes
covid_paises = covid_df.groupby('Country/Region', as_index=False).agg(Deaths=('Deaths', 'sum'))

# d) Mesclar com dados populacionais/continente
df_merge = covid_paises.merge(pop_df, on='Country/Region', how='left')

# e+f+g) Agrupar por continente, calcular mortes por milhão e ranquear
ranking_cont = (
    df_merge.groupby('Continent', as_index=False)
    .agg(TotalDeaths=('Deaths', 'sum'), TotalPop=('Population', 'sum'))
)
ranking_cont['Deaths_per_million'] = ranking_cont['TotalDeaths'] / ranking_cont['TotalPop'] * 1e6
ranking_cont = ranking_cont.sort_values('Deaths_per_million', ascending=False).reset_index(drop=True)
ranking_cont['Rank'] = ranking_cont.index + 1

print(ranking_cont[['Rank', 'Continent', 'Deaths_per_million']])
```
