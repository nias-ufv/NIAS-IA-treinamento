 **happy_words = ['great', 'excited', 'happy', 'nice', 'wonderful', 'amazing', 'good', 'best']

 **● Dicas:
 ○ apartir de uma variável inicialmente falsa (“is_happy = False”, por exemplo),

 ○ utilize um loop para analisar a frase completa, lembre-se que strings se

 **3. A partir do item “b)”, crie uma nova variável que conte quantos tweets felizes estão
 contidos na lista “tweets”, crie também outra variável que conte o número de tweets

 ● Dica: Serão utilizados loops dentro de loops para a realização deste item.**

```python
#Exercícios 1 e 3
# Listas de palavras felizes e tristes
happy_words = ['great', 'excited', 'happy', 'nice', 'wonderful', 'amazing', 'good', 'best']
sad_words = ['sad', 'bad', 'tragic', 'unhappy', 'worst']

# Lista de tweets
tweets = [
    "Wow, what a great day today!! #sunshine",
    "I feel sad about the things going on around us. #covid19",
    "I'm really excited to learn Python with @JovianML #zerotopandas",
    "This is a really nice song. #linkinpark",
    "The python programming language is useful for data science",
    "Why do bad things happen to me?",
    "Apple announces the release of the new iPhone 12. Fans are excited.",
    "Spent my day with family!! #happy",
    "Check out my blog post on common string operations in Python. #zerotopandas",
    "Freecodecamp has great coding tutorials. #skillup"
]

# Função para identificar se o tweet é feliz ou não
def identificar_emocao(tweet):
    # Convertendo o tweet para minúsculas e separando as palavras
    tweet = tweet.lower().split()
    # Verificando se há palavras felizes ou tristes no tweet
    feliz = any(palavra in tweet for palavra in happy_words)
    triste = any(palavra in tweet for palavra in sad_words)
    # Decidindo a emoção com base nas palavras encontradas
    if feliz:
        return "Feliz"
    elif triste:
        return "Triste"
    else:
        return "Neutro"

# Variáveis para contar o número de tweets felizes e tristes
tweets_felizes = 0
tweets_tristes = 0

# Loop para analisar todos os tweets e contar os felizes e tristes
for i, tweet in enumerate(tweets):
    emocao = identificar_emocao(tweet)
    if emocao == "Feliz":
        tweets_felizes += 1
    elif emocao == "Triste":
        tweets_tristes += 1
total_de_tweets = tweets_felizes + tweets_tristes
# Exibindo os resultados
print(f"Total de tweets felizes: {tweets_felizes}")
print(f"Total de tweets tristes: {tweets_tristes}")
print(f"Total de tweets: {total_de_tweets}")


```

```python
# Exercicio 2:

# Pegando o primeiro tweet
primeiro_tweet = tweets[0].lower()

# Variável inicial
is_happy = False

# Loop para verificar palavras felizes
for palavra in happy_words:
    if palavra in primeiro_tweet:
        is_happy = True
        break  # já achou, pode parar

# Mostrando o resultado
if is_happy:
    print("O primeiro tweet é FELIZ")
else:
    print("O primeiro tweet NÃO é feliz")
```
